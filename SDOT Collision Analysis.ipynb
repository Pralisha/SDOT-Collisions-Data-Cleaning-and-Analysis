<h3>Hypothesis testing on target features using chi-square tests </h3>

<p>H0: There is no relationship between driving under influence and collision <br>
H1: There is a strong relationship between driving under influence and collision <br>

chisquare test () :<br>

observe p value<br>
if pvalue < 0.05 -> Accept  alternate hypotheses<br>
else accept H0</p>

```python
from scipy.stats import chi2_contingency
```

```python
# Create contingency table
contingency_table = pd.crosstab(df['UNDERINFL'], df['SEVERITYDESC'])
```

```python
# Perform Chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

print("Chi-square statistic:", chi2)
print("p-value:", p)
print("Degrees of freedom:", dof)
```

<p><b>With a p-value of approximately 4.9594606070809714e-203, it is essentially zero, indicating a strong statistical significance. Since p < 0.05, we reject the null hypothesis and conclude that there is a significant association between 'UNDERINFL' and 'SEVERITYDESC' in the dataset.</b></p>

<h3>Test for normality using Kolmogorov-Smirnov test.</h3>

<h4>Kolmogorov Smirnov Test</h4>
<p><b>Null Hypothesis (H0):</b> The null hypothesis assumes that the sample data follows a normal distribution. 
<br>
<b>Alternative Hypothesis (H1):</b> The alternative hypothesis contradicts the null hypothesis. It suggests that the sample data does not follow a normal distribution.</p>
<p>Interpreting the results:<br>
alpha = 0.05  (Significance level)<br>
<br>
if pvalue > alpha:<br>
    print("Sample looks Gaussian (fail to reject H0)")<br>
  <br>  
else:<br>
    print("Sample does not look Gaussian (reject H0)")<br>
</p>

```python
# Let us make the severity code column numeric
df['SEVERITYCODE'] = df['SEVERITYCODE'].replace({'2b':2.5,'0':0,'1':1,'2':2,'3':3})
```

```python
# df['SEVERITYCODE'].dtype
```

```python
# print(df['SEVERITYCODE'].unique())
```

```python
from scipy.stats import kstest
# Extract the numerical column for testing
severity_code = df['SEVERITYCODE']

# Perform Kolmogorov-Smirnov test
statistic, p_value = kstest(severity_code, 'norm')


print("Kolmogorov-Smirnov statistic:", statistic)
print("p-value:", p_value)
```

```python
alpha=0.05
if p_value > alpha:
    print("Sample looks Gaussian (fail to reject H0)")
   
else:
    print("Sample does not look Gaussian (reject H0)")
```

<li>A p-value of 0.0 indicates that it is extremely unlikely that the sample data follows a normal distribution. Therefore, based on this result, we reject the null hypothesis and conclude that the sample data does not follow a normal distribution.</li>

<li>In summary, the KS test provides evidence that the distribution of the 'SEVERITYCODE' column in the SDOT dataset does not follow a normal distribution.</li>
